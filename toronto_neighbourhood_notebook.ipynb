Importing necessary packages initially

```python
import pandas as pd
import numpy as np
```

```python
from bs4 import BeautifulSoup
```

 Importing data from webpage and converting into dataframe

```python
data=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M',header=0)
```

```python
data2=data[0]
```

```python
df=pd.DataFrame(data2)
```

```python
df.head()
```

```python
df.columns=['Postal Code', 'Borough', 'Neighbourhood']
```

```python
df.head()
```

```python
df['Borough'].value_counts()
```

```python
df['Neighbourhood'].value_counts()
```

Removing rows that have Borough name "Not assigned"

```python
for x in range(df.shape[0]):
    if df['Borough'][x] == "Not assigned":
        df.drop(index=x,inplace=True)
```

```python
df=df.reset_index(drop=True)
```

```python
df['Latitude']=0.0
```

```python
df['Longitude']=0.0
```
