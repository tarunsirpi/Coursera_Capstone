##### Returning only relevant information for each nearby venue

(function defined from New York Neighbourhood Segmentation - Coursera - Applied Data Science Capstone)

```python
LIMIT=100
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues=getNearbyVenues(df['Neighbourhood'],df['Latitude'],df['Longitude'],radius=500)
```

```python
toronto_encoded= pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

toronto_encoded['Neighbourhood'] = toronto_venues['Neighbourhood']
altered_columns=[toronto_encoded.columns[-1]] + list(toronto_encoded.columns[:-1])
toronto_encoded=toronto_encoded[altered_columns]
toronto_encoded.head()
```

```python
toronto_encoded.shape
```

##### Getting the venues in each neighbourhood

```python
toronto_grouped=toronto_encoded.groupby('Neighbourhood').mean().reset_index()
toronto_grouped.head()
```

```python
toronto_grouped.shape
```

##### Getting the most common venues in each neighbourhoods

```python
for nei in toronto_grouped['Neighbourhood']:
    print(nei)
    temp_df = toronto_grouped[toronto_grouped['Neighbourhood'] == nei].T.reset_index()
    temp_df=temp_df.iloc[1:]
    temp_df.columns = ['Venue','Frequency']
    temp_df['Frequency'] = temp_df['Frequency'].astype(float)
    temp_df=temp_df.round({'Frequency':2})
    print(temp_df.sort_values('Frequency',ascending = False).reset_index(drop = True).head())
    
    print('\n')
    
```

###### Creating new dataframe with the most frequent venues of the neighbourhoods

```python
def high_freq_venues(row_categ,n):
    row_categ_sorted = row_categ.sort_values(ascending=False)
    return row_categ_sorted.index.values[0:n]
```

```python
n=5
title_indicators = ['st', 'nd', 'rd','th','th']
cols=['Neighbourhood']

for i in range(n):
    cols.append('{}{} most common venue'.format(i+1,title_indicators[i]))

toronto_freq_venues=pd.DataFrame(columns=cols)
toronto_freq_venues['Neighbourhood']=toronto_grouped['Neighbourhood']

for i in range(toronto_grouped.shape[0]):
    toronto_freq_venues.iloc[i , 1:]=high_freq_venues(toronto_grouped.iloc[i , 1:] , n)
    
```

##### Clustering

```python
from sklearn.cluster import KMeans
```

```python
k=5

toronto_clustering=toronto_grouped.drop('Neighbourhood',1)
model= KMeans(n_clusters=k,random_state=0).fit(toronto_clustering)
model.labels_
```

```python
toronto_freq_venues.insert(0,'Cluster labels',model.labels_)
```

```python
toronto_merged = df

toronto_merged = toronto_merged.join(toronto_freq_venues.set_index('Neighbourhood'),on='Neighbourhood')
toronto_merged
```
