(function defined from New York Neighbourhood Segmentation - Coursera - Applied Data Science Capstone)

```python
CLIENT_ID = 'NRM4SV1JOICSK2EPHQXC1H2BLLTYMII1ZAFAVK1QOMXDNTM2'
CLIENT_SECRET = '2FYSGCITV2J5JG51A3IFG3GBIPNJ4ULMJCH32S3WKO5OWHQH'
VERSION = '20200907'
```

```python
LIMIT=100
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&query=food&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Stadium', 
                  'Stadium Latitude', 
                  'Stadium Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

##### Getting the food venues with high frequency

```python
for sta in eng_food_grouped['Stadium']:
    print(sta)
    temp_df = eng_food_grouped[eng_food_grouped['Stadium'] == sta].T.reset_index()
    temp_df=temp_df.iloc[1:]
    temp_df.columns = ['Venue','Frequency']
    temp_df['Frequency'] = temp_df['Frequency'].astype(float)
    temp_df=temp_df.round({'Frequency':2})
    print(temp_df.sort_values('Frequency',ascending = False).reset_index(drop = True).head())
    
    print('\n')
```

```python
def high_freq_venues(row_categ,n):
    row_categ_sorted = row_categ.sort_values(ascending=False)
    return row_categ_sorted.index.values[0:n]
```

```python
n=5
title_indicators = ['st', 'nd', 'rd','th','th']
cols=['Stadium']

for i in range(n):
    cols.append('{}{} most common venue'.format(i+1,title_indicators[i]))

eng_freq_venues=pd.DataFrame(columns=cols)
eng_freq_venues['Stadium']=eng_food_grouped['Stadium']

for i in range(eng_food_grouped.shape[0]):
    eng_freq_venues.iloc[i , 1:]=high_freq_venues(eng_food_grouped.iloc[i , 1:] , n)
    
```
